```python
df_attrition= df[df['Attrition']=='Yes']
df_attrition.head()
```

```python
def calculate_attrition_rate(df, column):
    attrition_counts=df.groupby([column,'Attrition']).size().unstack(fill_value=0)
    attrition_rate= attrition_counts['Yes']/attrition_counts.sum(axis=1)*100
    attrition_rate_df=attrition_rate.reset_index()
    attrition_rate_df.columns= [column, 'AttritionRate']
    return attrition_rate_df
```

```python
fig , axes= plt.subplots(nrows=1,ncols=2, figsize=(15,6))

sns.kdeplot(data=df_attrition, x="Age",fill =True , ax=axes[0])
axes[0].set_title('Attretion by age')
axes[0].set_xlabel('Age')
axes[0].set_ylabel("destiny")

attrition_rate_df= calculate_attrition_rate(df,"Gender")

sns.barplot(data=attrition_rate_df, x="Gender",y="AttritionRate", ax=axes[1])
axes[1].set_title(f'Attrition Rate by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Attrition Rate (%)')
plt.tight_layout()
plt.show()
```

```python

df.columns = df.columns.str.strip()


attrition_rate = (df['Attrition'].value_counts(normalize=True)['Yes']) * 100

summary = pd.DataFrame({
    "Total_Employees": [len(df)],
    "Attrition_Rate_Percent": [round(attrition_rate, 2)],
    "Average_Age": [round(df['Age'].mean(), 2)],
    "Average_Monthly_Income": [round(df['MonthlyIncome'].mean(), 2)],
    "Department_Highest_Attrition": [
        df[df['Attrition'] == 'Yes'].groupby('Department')['Attrition'].count().idxmax()
    ],
    "JobRole_Highest_Attrition": [
        df[df['Attrition'] == 'Yes'].groupby('JobRole')['Attrition'].count().idxmax()
    ],
    "OverTime_Impact": [
        df[df['Attrition'] == 'Yes']['OverTime'].value_counts().idxmax()
    ]
})

summary.to_csv("../result/IBM_HR_Attrition_Summary.csv", index=False)

print("HR Attrition result file created successfully!")
```
